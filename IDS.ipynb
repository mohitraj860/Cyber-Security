```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("DataSet.csv")
df.head()
```

```python
df.describe()
```

```python
type(df)
```

```python
row,col=df.shape
unique=list(df['protocol_type'].unique())
print(unique)
for i in range(row):
    index=unique.index(df.iloc[i,1])
    df.iloc[i,1]=index
unique=list(df['service'].unique())
for i in range(row):
    index=unique.index(df.iloc[i,2])
    df.iloc[i,2]=index
unique=list(df['flag'].unique())
for i in range(row):
    index=unique.index(df.iloc[i,3])
    df.iloc[i,3]=index
df.head()
```

```python
row,col=df.shape
print("Dimension of data: row= ",row,"col= ",col)
print(df.columns)
```

```python
df_new=df[['protocol_type','service','flag','level']]
fig=plt.figure(figsize=(15,15),dpi=100)
sns.pairplot(df_new,hue="level")
plt.show()
plt.savefig("distribution.png")
```

```python
df1=df.drop(['level'],axis=1)
fig=plt.figure(figsize=(15,15),dpi=100)
sns.heatmap(df1.corr())
plt.show()
plt.savefig("correlation.png")
```
